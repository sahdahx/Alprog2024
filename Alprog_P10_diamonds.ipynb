## **Langkah 1: Menyiapkan Data**

Data bertipe .csv dapat diperoleh dari direktori/komputer maupun langsung diunduh dari internet. Pada pertemuan ini dicoba menggunakan data Diamond yang salah satunya tersedia pada situs https://www.kaggle.com/shivam2503/diamonds. Dataset ini memuat 11 variabel terkait 54.000 berlian (diamond) sebagai berikut.

- **price** price in US dollars (\$326--\$18,823)
***carat** weight of the diamond (0.2--5.01)
***cut** quality of the cut (Fair, Good, Very Good, Premium, Ideal)
***color** diamond colour, from J (worst) to D (best)
***clarity** a measurement of how clear the diamond is (I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF
(best))
***x** length in mm (0--10.74)
***y** width in mm (0--58.9)
***z** depth in mm (0--31.8)
***depth** total depth percentage = z / mean(x, y) = 2 * z / (x + y) (43--79)
***table** width of top of diamond relative to widest point (43--95)


```python
import pandas as pd
import seaborn as sb

dia_url = "https://raw.githubusercontent.com/TrainingByPackt/Interactive-Data-Visualization-with-Python/master/datasets/diamonds.csv"
diamond = pd.read_csv(dia_url)

# Untuk mengamati beberapa data teratas
x = diamond.head()
print(x)
```

```python
# Untuk mengamati beberapa data terbawah
y = diamond.tail()
print(y)
```

```python
# Untuk menyajikan statistika deskriptif masing-masing kolom dapat digunakan
z = diamond.describe()
print(z)
```

```python
# Kita juga dapat membentuk data frame baru yang memuat data dengan kriteria tertentu. Sebagai contoh, ingin dibentuk dataframe diamond_ideal yang hanya memuat data dari berlian yang memiliki cut bertipe ideal.
cut = diamond_ideal = diamond.loc[diamond['cut']=='Ideal']
print(cut)
```

## **Langkah 2: Menyajikan Histogram**

```python
# menyajikan histogram untuk variabel ‘carat’
diamond.hist(column='carat')
```

```python
sb.distplot(diamond.carat)
```

```python
# Perintah tersebut akan menghasilkan histogram yang dilengkapi dengan estimasi kernel (kernel density estimation) atau ‘poligon frekuensi’. Untuk menghilangkan estimasi tersebut, perintah di atas dapat dimodifikasi sebagai berikut.
sb.distplot(diamond.carat, kde = False)
```

## **Langkah 3: Menyajikan Diagram Batang **

```python
# Akan dibentuk diagram batang berdasar macam-macam potongan berlian (cut).
cut_macam = pd.crosstab(index=diamond['cut'],columns='count')
print(cut_macam)
cut_macam.plot(kind = 'bar')
```

```python
sb.catplot(x="cut", data=diamond, aspect=1.5, kind="count", color="g")
```

```python
import numpy
sb.set(style="whitegrid")
ax = sb.barplot(x="cut", y="price", data=diamond, estimator=numpy.mean)
```

```python
# Kita juga dapat menyajikan diagram batang yang lebih kompleks, misalnya dengan menambahkan pengkategorian pada masing-masing contoh. Sebagai contoh, kita inginmemodifikasi diagram di atas dengan menambahkan warna berlian.
ax = sb.barplot(x="cut", y="price", hue='color', data=diamond)
```

```python
# Untuk memperbaiki letak legenda/keterangan, perintah di atas dapat dimodifikasi menjadi
ax = sb.barplot(x='cut', y='price', hue='color', data=diamond)
ax.legend(loc='upper right',ncol=4)
ax = sb.barplot(x='cut', y='price', hue='color', data=diamond)
ax.legend(loc='upper right', ncol=4)
ax.set_xlabel('Jenis potongan', fontdict={'fontsize' : 14})
ax.set_ylabel('Harga', fontdict={'fontsize' : 14})
```

## **Langkah 4: Menyajikan Diagram Kotak-Garis dan Diagram Biola**

Untuk data bertipe interval/rasio, penyajian dalam bentuk diagram batang dirasa kurang informatif, terlebih bila akan dilakukan pembandingan data antar kelompok. Diperlukan sajian data yang langsung menunjukkan distribusi atau sebaran data.

Catplot (saya ragu untuk menyebut ‘diagram kucing’) memiliki bentuk seperti diagram titik (dot plot) namun diletakkan pada sumbu vertikal. Semakin tinggi kerapatan
titik menunjukkan semakin banyak data yang memilki nilai tersebut. Diagram ini dapat menunjukkan sebaran data, dan dibuat dengan perintah berikut.


```python
sb.catplot(x="cut", y="price", data=diamond)
```

```python
# Pada diagram kotak-garis atau boxplot, sumbu Y berupa variabel berskala interval/rasio, sedangkan sumbu X berupa variabel lain bertipe kategorik atau berskala nominal.
sb.boxplot(x="cut", y="price", data=diamond)
```

```python
# Diagram ini juga dapat dimodifikasi dengan menambahkan variabel lain sebagai hue.
sb.boxplot(x="cut", y="price", hue='color', data=diamond)
```

```python
sb.violinplot(x="cut", y="price", data=diamond)
```

## **Langkah 5: Menyajikan Diagram Pencar**

```python
# Membentuk diagram pencar untuk melihat hubungan antara “carat” dengan harga
sb.scatterplot(x="carat", y="price", data=diamond)
```

```python
# hue dapat digunakan untuk menambahkan variabel yang direpresentasikan dalam bentuk titik-titik yang berbeda. col dapat digunakan untuk memisahkan diagram pencar menurut variabel tertentu.
sb.scatterplot(x="carat", y="price", hue = "cut", data=diamond)
```

```python
# Bila perintah scatterplot diganti lmplot, akan diperoleh diagram pencar yang dilengkapi dengan estimasi garis regresi linear.
sb.lmplot(x="carat", y="price", data=diamond)
```

```python
# Untuk menghindari titik-titik yang bertumpuk terlalu banyak pada diagram pencar, dapat digunakan diagram heksagonal (hexagonal binning plot). Diagram ini dapat dibuat sebagai berikut.
import seaborn as sb
import matplotlib.pyplot as plt

sb.set(style="ticks")
sb.jointplot(x="carat", y="price", data=diamond, kind="hex", color="#4CB391")
plt.show()
```

## **Langkah 6: Menyajikan Diagram Lingkaran**

## **Langkah 7: Mengenal Matplotlib**
Diagram-diagram pada contoh di atas merupakan diagram-diagram ‘standar’ yang mudah dibuat dengan library pandas maupun seaborn. Diagram tersebut masih dapat  dimodifikasi dengan berbagai cara. Matplotlib merupakan library dasar visualisasi data  yang dapat dimanfaatkan untuk membuat diagram maupun memodifikasi diagram yang
telah dibuat sebelumnya. Perhatikan contoh berikut.

```python
import seaborn as sb
import matplotlib.pyplot as plt

# Membuat plot dasar
sb.lmplot(x="carat", y="price", data=diamond)

# Menambahkan keterangan dan anak panah dengan Matplotlib
plt.annotate(
    'Ini hasil yang menarik!', xy=(4.2, 20000), xytext=(2.5, 22000),
    arrowprops=dict(facecolor='red', shrink=0.05)
)

# Menampilkan plot
plt.show()

print("plt.annotate digunakan untuk menambahkan anotasi pada plot.")
print("'Ini hasil yang menarik!' adalah teks anotasinya.")
print("xy=(4.2, 20000) menentukan titik yang akan dianotasi.")
print("xytext=(2.5, 22000) menentukan posisi teks anotasi.")
print("arrowprops=dict(facecolor='red', shrink=0.05) mengatur tampilan anak panah (warna merah dan shrink=0.05).")
```

```python
# Contoh modifikasi lain dengan matplotlib

# Membuat plot dasar
sb.lmplot(x="carat", y="price", data=diamond)
# Menambahkan garis vertikal berwarna merah pada x = 1.5
plt.axvline(1.5, color='red')
# Menambahkan garis horizontal berwarna hijau pada y = 12000
plt.axhline(12000, color='green')
# Menampilkan plot
plt.show()
```
