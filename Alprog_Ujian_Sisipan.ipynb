# **Algoritma dan Pemrogaman untuk Statistika, 16 Maret 2024**

```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi y = 2 - |x|
def f(x):
    return 2 - np.abs(x)

# Range nilai x
x = np.linspace(-2, 2, 400)

# Plot fungsi y = 2 - |x|
plt.plot(x, f(x), 'r', label='$y = 2 - |x|$')

# Area yang dibatasi oleh kurva dan sumbu x
plt.fill_between(x, f(x), color='lightgray', alpha=0.5)

# Label sumbu dan judul
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Daerah yang Dibatasi oleh Kurva $y = 2 - |x|$ dan Sumbu-X')

# Tampilkan grid, legend, dan plot
plt.grid(True)
plt.legend()
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()
```
