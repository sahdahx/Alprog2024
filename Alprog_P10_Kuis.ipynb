Penyakit jantung merupakan salah satu permasalahan kesehatan yang penting. Dataset heart disease memuat sejumlah variabel yang diharapkan dapat digunakan untuk memprediksi penyakit jantung.

1. #3 (age) : usia pasien dalam tahun
2. #4 (sex) :  jenis kelamin pasien, 1 = laki-laki; 0 = perempuan.
3. #9 (cp) : tipe nyeri dada, 1 = typical angina; 2 = atypical angina; 3 = non-anginal pain; 4: asymptomatic.
4. #10 (trestbps) : tekanan darah pada saat pasien tiba di rumah sakit
5. #12 (chol) :  kadar kolesterol serum (mg/dl)
6. #16 (fbs) : kadar gula darah puasa > 120 mg/dl, 1 = ya; 0 = tidak.
7. #19 (restecg) : hasil elektrokardiograf saat istirahat, 0 = normal; 1 = abnormalitas pada ST-T; 2 = gejala hipertrofi pada ventrikel kiri.
8. #32 (thalach) :  ketercapaian denyut jantung maksimum
9. #38 (exang) : angina yang dipicu oleh aktivitas fisik, 1 = ya; 0 = tidak
10. #40 (oldpeak) : depresi ST pada elektrokardiograf saat aktivitas dibandingkan saat istirahat
11. #41 (slope) : pola kemiringan pada gelombang ST, 1 = upsloping, 2 = flat, 3 = downsloping
12. #44 (ca) :  banyaknya pembuluh darah (0-3) terwarnai saat fluoroscopy
13. #51 (thal) :  3 = normal; 6 = fixed defect; 7 = reversable defect
14. #58 (num) : diagnosis penyakit jantung berdasar angiografi, 0 = <50% diameter pembuluh darah mengalami penyempitan; 1= >50% diameter pembuluh darah mengalami penyempitan.

**1. Mengunduh data heart disease dan membaca data**

```python
# Membaca data
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt

url = "https://archive.ics.uci.edu/static/public/45/data.csv"
data = pd.read_csv(url)
print(data)
```

**2. Menuliskan program Python untuk membuka data tersebut, serta menampilkan data 5 pasien pertama dan 5 pasien terakhir.**

```python
print("menampilkan data 5 pasien pertama:")
print(data.head())
```

```python
print("menampilkan data 5 pasien terakhir:")
print(data.tail())
```

**3. Membuat sekurang-kurangnya dua bentuk visualisasi yang bermakna dari data tersebut dengan pustaka matplotlib, pandas, dan sebagainya.**

**4. Menginterpretasikan hasil visualisasi data yang dibuat pada langkah (3).**

```python
# Membuat DataFrame
df = pd.DataFrame(data)

# Membuat diagram garis
plt.figure(figsize=(14, 8))

plt.plot(df['age'], df['trestbps'], label='Tekanan Darah (trestbps)', marker='o')
plt.plot(df['age'], df['thalach'], label='Denyut Jantung Maksimum (thalach)', marker='d')

# Menambahkan judul dan label
plt.title('Diagram Garis Variabel Kesehatan pada Dataset Penyakit Jantung')
plt.xlabel('Usia Pasien (age)')
plt.ylabel('Nilai Variabel')
plt.grid(True)
plt.legend()

# Menampilkan diagram garis
plt.show()
```

# Visualisasi Data dengan Diagram Garis

### Interpretasi

1. **Tekanan Darah (trestbps)**:
   - Berdasarkan pola pada diagram garis tersebut, tekanan darah cenderung lebih tinggi pada pasien yang lebih tua. Misalnya, pasien dengan usia 67 tahun memiliki tekanan darah sebesar 160, sedangkan pasien yang lebih muda seperti usia 37 tahun memiliki tekanan darah sebesar 130.
   - Terdapat berbagai variasi, tetapi sebagian besar pasien yang lebih tua memiliki tekanan darah yang lebih tinggi dibandingkan dengan pasien yang lebih muda.

2. **Denyut Jantung Maksimum (thalach)**:
   - Berdasarkan pola pada diagram garis tersebut, denyut jantung maksimum cenderung lebih tinggi pada pasien yang lebih muda dan menurun seiring bertambahnya usia.
   - Contohnya, pasien usia 37 tahun memiliki denyut jantung maksimum sebesar 187, sementara pasien usia 67 tahun memiliki denyut jantung maksimum sebesar 108.

### Kesimpulan

- **Tekanan Darah (trestbps)** dan **Denyut Jantung Maksimum (thalach)** menunjukkan pola yang lebih jelas terkait dengan usia, di mana tekanan darah cenderung meningkat dan denyut jantung maksimum cenderung menurun seiring bertambahnya usia.

```python
# Membuat DataFrame
df = pd.DataFrame(data)

# Mengelompokkan data berdasarkan usia dan menghitung jumlah kasus penyakit jantung untuk setiap kelompok usia
age_groups = df.groupby('age')['num'].sum()

# Membuat diagram batang
plt.figure(figsize=(14, 8))
age_groups.plot(kind='bar', color='skyblue')

# Menambahkan judul dan label
plt.title('Diagram Batang Jumlah Kasus Penyakit Jantung Berdasarkan Usia')
plt.xlabel('Usia Pasien (age)')
plt.ylabel('Jumlah Kasus Penyakit Jantung')
plt.grid(axis='y')

# Menampilkan diagram batang

plt.show()
```

# Visualisasi Data dengan Diagram Batang

### Interpretasi

1. **Jumlah Kasus Penyakit Jantung Berdasarkan Usia**:
   - Berdasarkan pola dari diagram batang tersebut, kita bisa melihat bagaimana jumlah kasus penyakit jantung bervariasi berdasarkan kelompok usia.
   - Untuk usia tertentu, misalnya, jumlah kasus penyakit jantung tertinggi terlihat pada kelompok usia 58 tahun.

2. **Pengaruh Umur**:
   - Pasien yang berusia lebih tua (50 tahun ke atas) cenderung memiliki lebih banyak kasus penyakit jantung dibandingkan dengan pasien yang lebih muda (50 tahun ke bawah).
   - Tren ini menunjukkan bahwa risiko penyakit jantung cenderung meningkat seiring bertambahnya usia.

### Kesimpulan
- Berdasarkan diagram batang yang menunjukkan jumlah kasus penyakit jantung berdasarkan usia, dapat disimpulkan bahwa risiko penyakit jantung cenderung meningkat seiring bertambahnya usia, dengan jumlah kasus yang lebih tinggi pada pasien berusia 50 tahun ke atas dibandingkan dengan yang lebih muda.
